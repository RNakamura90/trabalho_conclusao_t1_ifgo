#### 2.3 Dificuldade de Identificação de Padrões

### 3. Metodologia e Escolha da Técnica de IA

Para solucionar o problema de interpretação dos fluxos financeiros presentes nas campanhas eleitorais, a metodologia deste trabalho afasta-se de abordagens preditivas tradicionais de Aprendizado de Máquina. O objetivo não é prever resultados eleitorais, estimar intenções de voto ou realizar classificações automáticas. O foco está na representação, exploração e interpretação de relações existentes em dados públicos já disponíveis.

Nesse contexto, a utilização de modelos estatísticos ou redes neurais não se mostra a alternativa mais adequada, uma vez que o problema central não envolve previsão, mas sim a identificação de conexões e caminhos relevantes entre diferentes entidades do ecossistema eleitoral. O desafio consiste em compreender como recursos financeiros percorrem uma rede composta por fontes de financiamento, partidos políticos, candidatos, tipos de despesas e setores econômicos fornecedores.

Por essa razão, a técnica de Inteligência Artificial selecionada foi a modelagem baseada em Grafos combinada com o Algoritmo de Dijkstra para cálculo de caminhos mínimos. De acordo com Russell e Norvig (2022), problemas que envolvem navegação, conectividade e descoberta de caminhos podem ser representados de forma natural por estruturas de grafos, nas quais entidades são modeladas como nós e seus relacionamentos como arestas.

A escolha se justifica porque os dados eleitorais analisados apresentam uma estrutura inerentemente relacional. Um recurso financeiro não existe de forma isolada; ele percorre diferentes entidades até atingir seu destino final. Dessa forma, representar essas relações como uma rede permite visualizar e analisar conexões que dificilmente seriam percebidas por meio de tabelas tradicionais ou consultas manuais.

Além disso, o Algoritmo de Dijkstra possibilita identificar os caminhos de menor custo entre dois pontos da rede. Neste trabalho, o conceito clássico de distância foi adaptado para representar a intensidade financeira dos relacionamentos observados. Assim, caminhos com maior volume financeiro agregado apresentam menor custo computacional, sendo interpretados como os fluxos economicamente mais relevantes dentro da rede eleitoral.

#### 3.1 Formalização do Modelo em Grafos

Para estruturar a rede de relacionamentos, os elementos presentes nas bases do Tribunal Superior Eleitoral (TSE) foram modelados como nós do grafo:

##### 3.1.1 Conjunto de Nós ($V$)

Os principais tipos de entidades considerados foram:

* Origens de Receita;
* Fontes de Receita;
* Partidos Políticos;
* Candidatos;
* Tipos de Despesa;
* Categorias Econômicas dos Fornecedores (CNAE).

Formalmente:

$$
V = {OrigemReceita, FonteReceita, Partido, Candidato, TipoDespesa, CNAEFornecedor}
$$

##### 3.1.2 Conjunto de Arestas ($E$)

As relações financeiras observadas nos dados foram modeladas como arestas direcionadas:

$$
OrigemReceita \rightarrow FonteReceita
$$

$$
FonteReceita \rightarrow Partido
$$

$$
Partido \rightarrow Candidato
$$

$$
Candidato \rightarrow TipoDespesa
$$

$$
TipoDespesa \rightarrow CNAEFornecedor
$$

Cada aresta representa um fluxo financeiro agregado identificado nas bases de receitas e despesas eleitorais.

##### 3.1.3 Definição dos Pesos

Para permitir a aplicação do Algoritmo de Dijkstra, cada aresta recebeu um peso associado ao volume financeiro movimentado.

Seja $v_{ij}$ o valor financeiro agregado entre os nós $i$ e $j$.

O custo da aresta foi definido como:

$$
w_{ij}=\frac{1}{1+\ln(1+v_{ij})}
$$

Essa transformação possui duas propriedades desejáveis:

* Valores financeiros elevados produzem custos menores;
* Diferenças extremas de escala são suavizadas pela função logarítmica.

Dessa forma, conexões economicamente mais relevantes tornam-se mais atrativas para o algoritmo de busca.

###### 3.1.4 Algoritmo de Dijkstra

Considere um grafo direcionado e ponderado:

$$
G=(V,E)
$$

Dado um nó de origem $s$ e um nó de destino $t$, o Algoritmo de Dijkstra busca o caminho cuja soma dos custos das arestas seja mínima:

$$
C(P)=\sum_{(i,j)\in P} w_{ij}
$$

onde $P$ representa um caminho possível entre os dois nós.

No contexto deste trabalho, isso significa identificar os fluxos financeiros mais relevantes entre diferentes componentes da rede eleitoral.

Por exemplo, um caminho identificado pelo algoritmo pode assumir a seguinte estrutura:

$$
Recursos\ de\ Pessoas\ Físicas
\rightarrow
Outros\ Recursos
\rightarrow
Partido
\rightarrow
Candidato
\rightarrow
Impulsionamento\ de\ Conteúdo
\rightarrow
Setor\ Econômico\ Fornecedor
$$

O custo total desse caminho representa uma medida agregada da intensidade financeira observada ao longo da rede.

#### 4.2 Definição do Problema de Busca

Para aplicação do Algoritmo de Dijkstra, o problema foi formalizado da seguinte forma:

##### 4.2.1 Estado Inicial

O estado inicial corresponde a uma origem de receita presente nas campanhas eleitorais, por exemplo:

$$

$$

ou

$$

##### 4.2.2 Espaço de Estados

O espaço de estados é composto por todos os nós existentes na rede financeira construída a partir dos dados eleitorais.

Formalmente:

$$
S = {Origens,\ Fontes,\ Partidos,\ Candidatos,\ Tipos\ de\ Despesa,\ CNAEs}
$$

##### 4.2.3 Ações (Transições)

Uma ação consiste em percorrer uma aresta válida do grafo.

Por exemplo:

$$
Recursos\ de\ Pessoas\ Físicas
\rightarrow
Outros\ Recursos
$$

ou

$$
Partido
\rightarrow
Candidato
$$

Cada transição representa uma relação financeira observada nos dados.

##### 4.2.4 Custo do Caminho

Para permitir a aplicação do algoritmo, os valores financeiros foram transformados em custos.

Seja $v_{ij}$ o valor financeiro agregado entre dois nós.

O peso da aresta foi definido por:

$$
w_{ij}=\frac{1}{1+\ln(1+v_{ij})}
$$

Essa transformação garante que conexões associadas a maiores volumes financeiros apresentem custos menores, tornando-as mais relevantes durante a busca.

O custo total de um caminho corresponde à soma dos pesos das arestas percorridas.

##### 4.2.5 Objetivo

O objetivo da busca consiste em encontrar o caminho de menor custo entre uma origem de receita e uma categoria econômica fornecedora.

No contexto deste trabalho, um caminho de menor custo representa um fluxo financeiro de maior relevância econômica dentro da rede eleitoral.

#### 4.3 Parametrização do Algoritmo

Essas funções permitiram transformar os dados brutos do TSE em uma representação estruturada baseada em grafos, possibilitando a aplicação do Algoritmo de Dijkstra e a interpretação visual dos principais fluxos financeiros observados nas campanhas eleitorais analisadas.



```python
# Importação de bibliotecas
from pathlib import Path
import sys
import requests
sys.path.append(str(Path("..").resolve()))
from src.tse_io import * #as principais funções foram importadas de outra pasta, porém vou apresenta-las aqui também
from src.graph_builder import *
import networkx as nx
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Função utilizada para baixar o arquivo zip do TSE, caso ele não exista localmente
def download_file(url: str, output_path: Path) -> None:
    output_path.parent.mkdir(parents=True, exist_ok=True)

    if output_path.exists():
        print(f"Arquivo já existe: {output_path}")
        return

    print("Baixando arquivo...")
    response = requests.get(url, timeout=120)
    response.raise_for_status()

    output_path.write_bytes(response.content)
    print(f"Arquivo salvo em: {output_path}")

# Função utilizada para ler um arquivo CSV específico de dentro do zip baixado do TSE    
def read_csv_from_zip(
    zip_path: Path,
    file_name: str,
    sep: str = ";",
    encoding: str = "latin1",
) -> pd.DataFrame:
    with zipfile.ZipFile(zip_path, "r") as z:
        if file_name not in z.namelist():
            raise FileNotFoundError(f"Arquivo não encontrado no ZIP: {file_name}")

        with z.open(file_name) as f:
            df = pd.read_csv(
                f,
                sep=sep,
                encoding=encoding,
                dtype=str,
                low_memory=False,
            )

    df["ARQUIVO_ORIGEM"] = file_name
    return df

# Função utilizada para listar os arquivos contidos no zip baixado do TSE    
def list_zip_files(zip_path: Path) -> list[str]:
    with zipfile.ZipFile(zip_path, "r") as z:
        return z.namelist()

# Função utilizada para ler o arquivo Brasil do zip baixado do TSE
def read_brasil_file(zip_path: Path, prefix: str) -> pd.DataFrame:
    file_name = f"{prefix}_BRASIL.csv"
    return read_csv_from_zip(zip_path, file_name)

# Função utilizada para filtrar os candidatos presidenciais do DataFrame (apenas os candidatos com os números 13 e 22 (Lula e Bolsonaro))
def filter_presidential_candidates(
    df: pd.DataFrame,
    candidate_numbers: list[str] | None = None,
) -> pd.DataFrame:
    if candidate_numbers is None:
        candidate_numbers = ["13", "22"]

    out = df.copy()

    if "DS_CARGO" in out.columns:
        out = out[
            normalize_text(out["DS_CARGO"]).eq("PRESIDENTE")
        ].copy()

    if "NR_CANDIDATO" in out.columns:
        out = out[
            out["NR_CANDIDATO"].isin(candidate_numbers)
        ].copy()

    return out

# Função parse para valores monetários, convertendo os valores de string para float, considerando o formato brasileiro
def parse_money(series: pd.Series) -> pd.Series:
    return (
        series.astype(str)
        .str.replace(".", "", regex=False)
        .str.replace(",", ".", regex=False)
        .astype(float)
    )
    
# Função para construção do grafo de categorias, utilizando as funções auxiliares para construir as arestas e agregá-las   
def build_category_graph(receitas_pres, despesas_pres):
    raw_edges = build_category_edges(receitas_pres, despesas_pres)
    edges = aggregate_edges(raw_edges)

    G = nx.DiGraph()

    for _, row in edges.iterrows():
        G.add_edge(
            row["source"],
            row["target"],
            tipo=row["tipo"],
            valor_total=row["valor_total"],
            qtd_registros=row["qtd_registros"],
            weight=row["weight"],
        )

    return G, edges

#Função para limpar os rótulos dos nós, removendo o prefixo de categoria (ex: "categoria::") para uma visualização mais limpa nos gráficos
def clean_label(node):
    return node.split("::", 1)[1] if "::" in node else node

#Função para extrair o subgrafo de um candidato específico, incluindo as arestas de entrada, saída, e as arestas conectadas aos nós vizinhos (upstream e downstream)
def get_candidate_subgraph_edges(edges, candidate_name):
    candidate_node = f"candidato::{candidate_name}"

    incoming = edges[
        edges["target"].eq(candidate_node)
    ].copy()

    outgoing = edges[
        edges["source"].eq(candidate_node)
    ].copy()

    previous_nodes = incoming["source"].unique()

    upstream = edges[
        edges["target"].isin(previous_nodes)
    ].copy()

    next_nodes = outgoing["target"].unique()

    downstream = edges[
        edges["source"].isin(next_nodes)
    ].copy()

    candidate_edges = pd.concat(
        [upstream, incoming, outgoing, downstream],
        ignore_index=True
    ).drop_duplicates()

    return candidate_edges

#Função para plotar o gráfico de Sankey para um candidato específico, utilizando o subgrafo extraído pela função anterior, e considerando apenas as top_n arestas mais relevantes
def plot_candidate_sankey(edges, candidate_name, top_n=40):
    df = get_candidate_subgraph_edges(edges, candidate_name)

    df = df.sort_values("valor_total", ascending=False).head(top_n)

    labels = pd.Index(
        pd.concat([df["source"], df["target"]]).unique()
    )

    node_id = {label: i for i, label in enumerate(labels)}

    fig = go.Figure(
        data=[
            go.Sankey(
                node=dict(
                    label=[clean_label(x) for x in labels],
                    pad=15,
                    thickness=15,
                ),
                link=dict(
                    source=df["source"].map(node_id),
                    target=df["target"].map(node_id),
                    value=df["valor_total"],
                    customdata=df["tipo"],
                    hovertemplate=(
                        "Origem: %{source.label}<br>"
                        "Destino: %{target.label}<br>"
                        "Valor: R$ %{value:,.2f}<br>"
                        "Tipo: %{customdata}<extra></extra>"
                    ),
                ),
            )
        ]
    )

    fig.update_layout(
        title_text=f"Fluxo agregado da campanha: {candidate_name}",
        font_size=10,
        height=700,
    )

    return fig

#Função para calcular o custo total de um caminho no grafo, somando os pesos das arestas ao longo do caminho
def path_cost(G, path):
    cost = 0

    for i in range(len(path) - 1):
        cost += G[path[i]][path[i + 1]]["weight"]

    return cost

#Função para encontrar os caminhos mais curtos (com menor custo) entre as origens de receita e os destinos de fornecedores, passando por um candidato específico
def candidate_paths(
    G,
    candidate_name,
):
    candidate_node = f"candidato::{candidate_name}"

    origens = [
        n for n in G.nodes
        if n.startswith("origem_receita::")
    ]

    destinos = [
        n for n in G.nodes
        if n.startswith("cnae_fornecedor::")
    ]

    rows = []

    for origem in origens:
        for destino in destinos:

            try:
                path = nx.shortest_path( #Dijkstra
                    G,
                    source=origem,
                    target=destino,
                    weight="weight",
                )

                if candidate_node not in path:
                    continue

                rows.append({
                    "origem": origem,
                    "destino": destino,
                    "candidato": candidate_name,
                    "custo": path_cost(G, path),
                    "caminho": " -> ".join(path),
                    "num_passos": len(path) - 1,
                })

            except nx.NetworkXNoPath:
                pass

    return pd.DataFrame(rows)

#Função para descrever um caminho específico no grafo, detalhando as arestas e seus atributos ao longo do caminho
def describe_path(G, path):
    rows = []

    for i in range(len(path) - 1):
        source = path[i]
        target = path[i + 1]
        edge = G[source][target]

        rows.append({
            "step": i + 1,
            "source": source,
            "target": target,
            "tipo": edge["tipo"],
            "valor_total": edge["valor_total"],
            "qtd_registros": edge["qtd_registros"],
            "weight": edge["weight"],
        })

    return pd.DataFrame(rows)
```

### Aplicação de Dijkstra

```python
paths_bolsonaro = candidate_paths(
    G_cat,
    "JAIR MESSIAS BOLSONARO"
)

paths_lula = candidate_paths(
    G_cat,
    "LUIZ INÁCIO LULA DA SILVA"
)

comparacao = pd.concat(
    [
        paths_lula,
        paths_bolsonaro,
    ],
    ignore_index=True
)

comparacao.groupby(
    "candidato",
    as_index=False
).agg(
    menor_custo=("custo", "min"),
    custo_medio=("custo", "mean"),
    qtd_caminhos=("custo", "count"),
)
```

Com o grafo construído, aplicou-se o Algoritmo de Dijkstra para identificar os caminhos de menor custo entre origens de receita e categorias econômicas fornecedoras. Como o custo das arestas foi definido de forma inversamente proporcional ao valor financeiro agregado, os caminhos mínimos representam os fluxos economicamente mais relevantes.

Os custos mínimos e médios dos caminhos são bastante semelhantes entre os dois candidatos, indicando que os fluxos financeiros mais intensos apresentam força agregada comparável. 

A principal diferença observada está na quantidade de caminhos identificados: a rede associada a Jair Bolsonaro apresentou maior número de conexões possíveis entre origens de receita e categorias econômicas fornecedoras, sugerindo maior diversidade relacional no grafo.

#### 6.1 Organização e Engenharia dos Dados de Entrada (X)

Os dados de entrada continuariam sendo obtidos a partir das bases públicas do Tribunal Superior Eleitoral (TSE), porém com uma abrangência maior, contemplando múltiplos ciclos eleitorais.

As principais variáveis coletadas poderiam incluir:

* Valores de receitas recebidas;
* Origens das receitas;
* Fontes de financiamento;
* Partidos políticos;
* Cargo disputado;
* Tipo de despesa realizada;
* Categoria econômica dos fornecedores (CNAE);
* Quantidade de fornecedores distintos;
* Quantidade de doadores distintos;
* Distribuição percentual dos gastos por categoria;
* Informações temporais da campanha.

Como a maior parte dessas informações possui natureza tabular, seria necessário realizar uma etapa de engenharia de atributos.

As variáveis numéricas seriam normalizadas utilizando técnicas como Min-Max Scaling ou Standard Scaling:

$$
x_{norm}=\frac{x-\mu}{\sigma}
$$

As variáveis categóricas, como partido político, origem de receita e CNAE, poderiam ser representadas por meio de embeddings aprendidos pela própria rede neural, permitindo capturar relações semânticas entre categorias.

O vetor final de entrada seria composto pela concatenação das variáveis numéricas normalizadas e dos embeddings das variáveis categóricas.

#### 6.2 Representação e Arquitetura Teórica da Rede

Como os dados possuem estrutura predominantemente tabular, uma arquitetura baseada em camadas densas (*Fully Connected Neural Networks*) seria a mais adequada.

Uma arquitetura hipotética poderia ser composta por:

* Camada de Entrada;
* Camada de Embedding para atributos categóricos;
* Camada Densa com 128 neurônios;
* Camada Densa com 64 neurônios;
* Camada Densa com 32 neurônios;
* Camada de Saída.

A função de ativação utilizada nas camadas ocultas seria a ReLU:

$$
ReLU(x)=max(0,x)
$$

A utilização da ReLU permite à rede capturar relações não lineares entre as variáveis financeiras, algo comum em sistemas complexos de financiamento eleitoral.

Para reduzir o risco de sobreajuste (*overfitting*), poderiam ser adicionadas camadas de Dropout entre as camadas densas.

#### 6.3 Estrutura da Saída Esperada (Y)

A estrutura da saída dependeria do objetivo de negócio definido para a solução.

Uma possibilidade seria formular o problema como uma classificação multiclasses, na qual a rede deveria identificar o perfil predominante de uma campanha eleitoral.

Exemplos de classes:

* Predominância de financiamento por pessoas físicas;
* Predominância de financiamento partidário;
* Predominância de Fundo Especial;
* Perfil financeiro híbrido.

Nesse cenário, a camada de saída utilizaria a função Softmax:

$$
Softmax(z_i)=\frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}}
$$

A saída da rede corresponderia a uma distribuição de probabilidades para cada perfil de campanha.

Como função de perda, seria utilizada a Categorical Cross Entropy:

$$
L=-\sum_{i=1}^{K} y_i\log(\hat{y}_i)
$$

onde:

* $y_i$ representa a classe real;
* $\hat{y}_i$ representa a probabilidade prevista pela rede.

O treinamento seria realizado utilizando o algoritmo Adam, amplamente empregado em problemas de classificação tabular devido à sua estabilidade e eficiência computacional.

A solução proposta mostrou-se viável para aplicações relacionadas à transparência pública e análise exploratória de dados eleitorais, apresentando como principal vantagem a sua capacidade de explicação.
